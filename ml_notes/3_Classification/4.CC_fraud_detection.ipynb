## Handling Class Imbalance: Undersampling, Oversampling, and SMOTE

### What is Class Imbalance?

**Class imbalance** occurs when one class significantly outnumbers the other in a dataset. In this credit card fraud dataset:
- ~99.8% of transactions are valid (Class 0)
- ~0.2% of transactions are fraudulent (Class 1)

This creates a problem: machine learning models tend to be biased toward the majority class (predicting "valid" for everything), achieving high accuracy without identifying fraud.

---

### Three Techniques to Handle Class Imbalance

#### 1. **Undersampling**
Remove samples from the majority class to balance the dataset.

**Pros:**
- Simple to implement
- Reduces training time (smaller dataset)

**Cons:**
- Loses valuable information from majority class
- May hurt model performance
- Only use if you have a very large dataset

**Example:**
```python
# Keep all 492 fraud cases, randomly sample 492 valid cases
valid_indices = data[data['Class'] == 0].index
fraud_indices = data[data['Class'] == 1].index
undersampled_valid = np.random.choice(valid_indices, len(fraud_indices), replace=False)
balanced_indices = np.concatenate([fraud_indices, undersampled_valid])
balanced_data = data.loc[balanced_indices]
```

---

#### 2. **Oversampling**
Duplicate samples from the minority class to balance the dataset.

**Pros:**
- Keeps all original data
- No information loss
- Simple to implement

**Cons:**
- Risk of overfitting (repeating exact same fraud cases)
- Increases training time (larger dataset)
- Model may memorize fraudulent patterns instead of generalizing

**Example:**
```python
# Duplicate fraud cases to match valid cases
fraud_indices = data[data['Class'] == 1].index
oversampled_fraud = np.random.choice(fraud_indices, len(valid_indices), replace=True)
balanced_indices = np.concatenate([valid_indices, oversampled_fraud])
balanced_data = data.loc[balanced_indices]
```

---

#### 3. **SMOTE (Synthetic Minority Over-sampling Technique)** ✅ Best Option

SMOTE creates **synthetic (artificial) fraud samples** instead of duplicating existing ones. It generates new minority class samples that are similar but not identical to real fraud cases.

**How SMOTE Works:**

```
Step 1: Pick a fraudulent transaction
        Example: [V1=1.2, V2=3.4, V3=0.8, ...]

Step 2: Find its k-nearest fraudulent neighbors
        Example: Find 5 similar fraud cases

Step 3: Create new synthetic fraud sample
        Blend the picked sample with a random neighbor
        New sample = Original + random_weight × (Neighbor - Original)
        Example: [V1=1.3, V2=3.5, V3=0.9, ...]

Step 4: Repeat until fraud cases match valid cases
```

**Pros:**
- ✅ Generates realistic synthetic minority samples
- ✅ No information loss (keeps all original data)
- ✅ Avoids overfitting better than simple oversampling
- ✅ Creates diverse fraud patterns (not just duplicates)

**Cons:**
- Slightly more complex
- Requires fitting on training set only (not test set)

**Important:** Always apply SMOTE ONLY on training data, not test data. The test set should remain imbalanced to reflect real-world conditions.

---

### When to Use Each Technique

| Technique | When to Use | Risk |
|-----------|------------|------|
| **Undersampling** | Very large dataset (millions of rows) | Loses important information |
| **Simple Oversampling** | Quick baseline, small projects | Overfitting risk |
| **SMOTE** | Most classification problems ✅ | Requires `imbalanced-learn` library |
| **Class Weights** | When you can't change data | Less intuitive |

---

### Summary

- **Class imbalance** = one class much larger than other (fraud detection, disease prediction, etc.)
- **Problem:** Model biases toward majority class, misses minority class
- **Solutions:** Undersampling, Oversampling, SMOTE
- **Best practice:** Use **SMOTE** on training data only to create synthetic minority samples
- **Remember:** Apply balancing to training set, evaluate on imbalanced test set

## Logistic Regression — Short Overview

**What it is:**
Logistic Regression is a linear model for binary classification that models the probability of the positive class using the logistic (sigmoid) function.

**Model form:**
- Compute linear score: z = w0 + w1·x1 + … + wn·xn
- Convert to probability: P(y=1|x) = 1 / (1 + exp(-z))

**Predictions:**
- `predict_proba(X)[:, 1]` → calibrated probabilities (useful for thresholding)
- `predict(X)` → class labels (0/1)
- `decision_function(X)` → raw decision scores

**Regularization:**
- Add `penalty='l1'` or `penalty='l2'` to avoid overfitting; tune `C` (inverse strength).

**Why use it here:**
- Fast to train and predict, outputs interpretable coefficients and probabilities, and works well after SMOTE balancing.

**Quick usage tip:**
Train on balanced (SMOTE) training data, use `predict_proba` to set an operational threshold, and evaluate on the original imbalanced test set.

````python
#Training a logistic regression model
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, auc, roc_curve
````

````python
parameters = {"penalty": ['l1', 'l2'], 'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
````

In your Logistic Regression code, **`penalty`** refers to **regularization** — a technique to prevent overfitting by penalizing the model for having large coefficients.

```python
parameters = {"penalty": ['l1', 'l2'], 'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
#           ↑ Testing two types of penalties
```

---

### What is Penalty/Regularization?

Without regularization, Logistic Regression tries to fit the training data perfectly, which can **overfit** (memorize noise instead of learning general patterns).

**Penalty adds a cost** for having large coefficient values, forcing the model to keep coefficients small and simpler.

---

### The Two Penalty Types

#### **L1 Penalty (Lasso)**
```python
"penalty": 'l1'
```

**How it works:**
- Adds the **sum of absolute values** of coefficients to the loss function
- Loss = Classification Error + C × (|w₁| + |w₂| + |w₃| + ... + |wₙ|)

**Effect:**
- Forces some coefficients to **exactly 0** (feature elimination)
- Performs **feature selection** automatically
- Result: Only the most important features are used

**Example:**
```
Before L1: w = [0.5, 0.3, 0.8, 0.1, 0.6]
After L1:  w = [0.4, 0.0, 0.7, 0.0, 0.5]
           ↑    Eliminated weak features
```

**Use when:**
- You want to know which features matter most
- You want a simpler, more interpretable model
- You have many irrelevant features

---

#### **L2 Penalty (Ridge)**
```python
"penalty": 'l2'
```

**How it works:**
- Adds the **sum of squared values** of coefficients to the loss function
- Loss = Classification Error + C × (w₁² + w₂² + w₃² + ... + wₙ²)

**Effect:**
- Shrinks coefficients toward 0 but never exactly to 0
- Keeps all features but with smaller weights
- Result: More balanced, smoother model

**Example:**
```
Before L2: w = [0.5, 0.3, 0.8, 0.1, 0.6]
After L2:  w = [0.3, 0.2, 0.5, 0.08, 0.4]
           ↑    All features reduced but kept
```

**Use when:**
- You want to keep all features
- You want a smoother, more stable model
- Features are highly correlated

---

### L1 vs L2 Visual Comparison

```
L1 (Lasso):           L2 (Ridge):
w₂                    w₂
│                     │
│  /\                 │    ←← Diamond shape
│ /  \  ← Diamond     │  /      (all features reduced)
│/    \               │ /
└──────── w₁          └──────── w₁

Preferred direction:  Preferred direction:
Push to ZERO          Push toward center
(feature selection)   (all features shrunk)
```

---

### In Your Code

```python
parameters = {"penalty": ['l1', 'l2'], 'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}

lr = LogisticRegression()
clf = GridSearchCV(lr, parameters, cv=5, verbose=5, n_jobs=3)
clf.fit(X_train_res, y_train_res)
```

**What's happening:**
1. GridSearchCV tests **2 penalties** × **7 C values** = **14 combinations**
2. Each combination is tested with **5-fold cross-validation**
3. Finds the best combination (e.g., "L2 penalty with C=10")
4. Reports which penalty and regularization strength works best for fraud detection

---

### Quick Summary

| Aspect | L1 | L2 |
|--------|-----|-----|
| **Formula** | Sum of absolute values | Sum of squared values |
| **Effect on coefficients** | Some → 0, others reduced | All reduced, none to 0 |
| **Feature selection** | ✅ Yes | ❌ No |
| **Use for fraud** | Better (simpler model) | Also works (more stable) |

**For your fraud detection:** L1 is slightly preferred because you want to identify which transaction features most strongly indicate fraud (interpretability).

````python
lr = LogisticRegression()
clf = GridSearchCV(lr, parameters, cv=5, verbose=5, n_jobs=3)
k = clf.fit(X_train_res, y_train_res)
````

````python
k.best_params_
````

````python
#Evaluate the model
lr_gridcv_best = clf.best_estimator_
y_test_pre = lr_gridcv_best.predict(X_test)
````

````python
cnf_matrix_test = confusion_matrix(y_test, y_test_pre)
print("Recall metric in the test dataset:", (cnf_matrix_test[1,1]/(cnf_matrix_test[1,0]+cnf_matrix_test[1,1] )))
````

````python
y_train_pre = lr_gridcv_best.predict(X_train_res)
cnf_matrix_train = confusion_matrix(y_train_res, y_train_pre)
print("Recall metric in the train dataset:", (cnf_matrix_train[1,1]/(cnf_matrix_train[1,0]+cnf_matrix_train[1,1] )))
````

````python
#Visualize conf matrix
from sklearn.metrics import ConfusionMatrixDisplay
#plot conf matrix on test data using predictions
class_names = ['Not Fraud', 'Fraud']
ConfusionMatrixDisplay.from_predictions(y_test, y_test_pre, display_labels=class_names, values_format='.6g')
plt.title("Test data Confusion Matrix")
plt.show()
````

````python
#Plot ROC Curve and check AUC
y_k = k.decision_function(X_test)
fpr, tpr, thresholds = roc_curve(y_test, y_k)
roc_auc = auc(fpr, tpr)
print("ROC-AUC:", roc_auc)
````

````python
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b',label='AUC = %0.3f'% roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.0])
plt.ylim([-0.1,1.01])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
````
