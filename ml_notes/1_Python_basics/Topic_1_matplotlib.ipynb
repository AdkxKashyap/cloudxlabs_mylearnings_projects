```python
import matplotlib
##Pyplot is a module of Matplotlib which provides simple functions to add plot elements like lines, images, text, etc. to the current axes in the current figure. 
import matplotlib.pyplot as plt
import numpy as np
matplotlib.__version__
```

```python
##Getting Stared with Matplotlib - Create a Simple Plot using Pyplot

X = np.array([1, 2, 5, 8, 12])
Y = np.array([3, 6, 7, 12, 9])
plt.plot(X, Y)
##use the show method to display this plot
plt.show()
```

```python
##Adding Title and Axis Label to Charts using Pyplot
X = np.array([1, 2, 5, 8, 12])
Y = np.array([3, 6, 7, 12, 9])
plt.plot(X, Y)
plt.title("This is the title")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

```python
##Creating Bar Graphs using Pyplot
subjects = ["Maths", "Biology", "Chemistry", "Physics", "English", "Computers"]
marks = [97, 68, 59, 81, 77, 92]
plt.bar(subjects, marks, color='blue')

plt.title("Bar Graph Example")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
##Creating Clustered Bar Graphs using Pyplot
"""
The clustered bar graph extends the standard bar 
graph from looking at numeric values across one categorical 
variable to two or more categorical variables. Each bar in a standard bar graph is divided into a number of sub-bars stacked side to side, each one 
corresponding to a level of the subsequent categorical variables.
"""
subjects = ["Maths", "Biology", "Chemistry", "Physics", "English", "Computers"]
student1 = [97, 68, 59, 81, 77, 92]
student2 = [88, 61, 80, 40, 62, 52]
student3 = [54, 62, 77, 54, 71, 98]

#index is x-axis coords. Eg-> for S1 coords for Maths will be (0,97) for S2 - (0.03,88).    
index = np.arange(6)
width = 0.03
plt.bar(index, student1, width, color="aqua", label="Student 1")
plt.bar(index+width, student2, width, color="green", label="Student 2")
plt.bar(index+2*width, student3, width, color="blue", label="Student 3")

plt.title("Stacked Bar Graph Example")
plt.xlabel("Students")
plt.ylabel("Marks")
##Since we used index above for x-axis, we need to set the ticks accordingly to show subject names.
plt.xticks(index+width/2, subjects)
plt.legend()
plt.show()
```

```python
#Creating Pie Charts using Pyplot
firms = ["Firm A", "Firm B", "Firm C", "Firm D", "Firm E"]
market_shares = [10, 40, 30, 5, 15]
##explode is used to focus on a particular data.Here we are focusing on firm with lowest market share
Explode = [0, 0, 0, 0.5, 0]
#Here startangle=45 means that everything is rotated counter-clockwise by 45 degrees.
plt.pie(market_shares, explode=Explode, labels=firms, shadow=True, startangle=45)

plt.axis('equal')
plt.legend(title = "List of Firms")
plt.show()
```

```python
#Creating Histograms using Pyplot
x = np.random.randn(10000)
plt.title("Histogram Example")
plt.xlabel("Random Data")
plt.ylabel("Frequency")
plt.hist(x, 10)
plt.show()
```

```python
#Create 3D Projections using Matplotlib
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x,y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```

```python
#Create a Scatter Plot using Pyplot
from numpy.random import rand
for color in ['red', 'green', 'blue']:
    n = 100
    #Generate random x, y, and scale values.Scale determines the size of each point in the scatter plot.rand() will return 2x100 array of random values between 0 and 1.x and y are 1st and 2nd rows of that array respectively.
    x, y = rand(2, n)
    scale = 500.0 * rand(n) ** 5
    #For each color this will plot n points with the specified color, size (scale), transparency (alpha), and edge color.
    plt.scatter(x, y, s=scale, c=color, alpha=0.3, edgecolors='blue')

plt.grid(True)
plt.show()
```
