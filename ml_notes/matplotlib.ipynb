```python
import matplotlib
##Pyplot is a module of Matplotlib which provides simple functions to add plot elements like lines, images, text, etc. to the current axes in the current figure. 
import matplotlib.pyplot as plt
import numpy as np
matplotlib.__version__
```

```python
##Getting Stared with Matplotlib - Create a Simple Plot using Pyplot

X = np.array([1, 2, 5, 8, 12])
Y = np.array([3, 6, 7, 12, 9])
plt.plot(X, Y)
##use the show method to display this plot
plt.show()
```

```python
##Adding Title and Axis Label to Charts using Pyplot
X = np.array([1, 2, 5, 8, 12])
Y = np.array([3, 6, 7, 12, 9])
plt.plot(X, Y)
plt.title("This is the title")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

```python
##Creating Bar Graphs using Pyplot
subjects = ["Maths", "Biology", "Chemistry", "Physics", "English", "Computers"]
marks = [97, 68, 59, 81, 77, 92]
plt.bar(subjects, marks, color='blue')

plt.title("Bar Graph Example")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```
